```python
# This function will print a description of the program. It will also explain how to signal the end of the set of data.
def program4_introduction():
    print("This program will read in three integers representing a student's scores on three tests."+
          " The program will determine if these three scores are valid."+
          " If the scores are valid, the program will classify the scores in various ways."+ 
          " The program will determine if three scores are same."+
          " The program will round the score and determine if rounded three scores are same."+ 
          " When the program process all dataset, it will display the number of dataset, number of valid dataset, and number of invalid dataset\n")
    
#the function aretheyvalid which will receive three integer parameters representing three test scores.
#parameters: x,y,z

def aretheyvalid(x,y,z):
    
    #initial bool variable 
    condition=False
    
    #if x and y and z are all (smaller or equal to 100) or (greater or euqual to 0)
    #the scores will be valid
    #then the bool variable will be true, means the scores are valid
    if (x in range (0,101)) and (y in range (0,101)) and (z in range (0,101)):
        condition=True
    
    #if x or y or z not in range of (0-100)
    #the bool variable will be false, means the scores are invalid
    elif (x not in range (0,101)) or (y not in range (0,101)) or (z not in range (0,101)):
        condition=False
        
        #if there is score invalid, the program will display outputs
        print("The set of data is invalid: ")
        
        #when x is invalid:
        if x <0:
            print(f"The score {x} is too small")
            
        elif x >100:
            print(f"The score {x} is too big")
        
        #when y is invalid:
        if y <0:
            print(f"The score {y} is too small")
            
        elif y >100:
            print(f"The score {y} is too big")
        
        #when z is invalid:
        if z <0:
            print(f"The score {z} is too small")
            
        elif z >100:
            print(f"The score {z} is too big")
    
    #return the bool varibale value
    return condition

#The function will compare three values and print one of the following four messages: 
#all three numbers are the same; 
#two numbers are the same and the third is larger (there are three ways this can happen); 
#two numbers are the same and the third is smaller (also three ways for this to occur); 
#all three numbers are different. 

#the function aretheysame which will receive three integer parameters

def aretheysame(x,y,z):
    #if the three scores are valid, the function will compare values
    if aretheyvalid(x,y,z)==True:
        print("Valid set of scores")
        #condition: all three numbers are the same; 
        if x==y==z:
            print("All three numbers are the same")
        
        #condition:two numbers are the same and the third is smaller
        elif (x==y and x>z) or (x==z and x>y) or (y==z and y>x):
            print("Two numbers are the same and the third is smaller")
        
        #condition:two numbers are the same and the third is larger    
        elif (x==y and x<z) or (x==z and x<y) or (y==z and y<x):
            print("Two numbers are the same and the third is larger")
        
        #condition: all three numbers are different
        elif x!=y and x!=z and y!=z:
            print("All three numbers are different")
            
#the function roundscore which will receive one integer parameter,
def roundscore(x):
    
    #convert integer to string
    num=str(x)
    #convert string to list
    num=list(num)
    
    #check the last number of list
    #if last number larger or euqal to 5,then round up the number
    #for example: 95, last number is 5, so round up to 100
    if int(num[-1])>=5:
        num_round=10-(int(num[-1]))
        x+=num_round
    
    #if last number smaller to 5, round down the number
    #for example: 94, last number is 4, so round down to 90
    elif int(num[-1])<5:
        x-=(int(num[-1]))
    
    #return rounded score
    return x
```

```python
#the main function
def program4_main():
    
    #call introduction function
    program4_introduction()
    
    #open the data file of student scores
    file=open('question4_data.txt','r')
    
    #read the first line
    line=file.readline().rstrip()
    
    #initial the varibales
    total_count=0
    valid_count=0
    invalid_count=0
    
    #when the line is not empty, which means there are scores in the line
    #the program will continue run
    while line!='':
        
        #convert the string in the line to the list
        line=line.split(',')
        x=int(line[0])
        y=int(line[1])
        z=int(line[2])
        
        #display scores
        print(f"The three scores are {x}   {y}   {z}")
        
        #check if the scores are valid
        if aretheyvalid(x,y,z)==True:
            
            #call aretheysame function
            aretheysame(x,y,z)
            
            #x,y,z variables, call roundscore function
            x=roundscore(x)
            y=roundscore(y)
            z=roundscore(z)
            
            #display outputs after score rounded
            print(f"The three scores after round are {x}    {y}    {z}")
            
            #call aretheysame function to check the rounded scores
            aretheysame(x,y,z)
            
            #read next line in the file
            line=file.readline().rstrip()
            
            #count the number of data group
            total_count+=1
            
            #count the number of valid group 
            valid_count+=1
            print('\n')
       
        #if there is a score not valid, read the next line
        else:
            line=file.readline().rstrip()
            
            #add the group to total number of groups
            total_count+=1
            
            #count the number of invalid group
            invalid_count+=1
            print('\n')
    
    #display outputs
    print(f"The total number of group processed: {total_count}")
    print(f"The total number of valid group: {valid_count}")
    print(f"The total number of invalid group: {invalid_count}")
    file.close()
    
#call main function
program4_main()
```
